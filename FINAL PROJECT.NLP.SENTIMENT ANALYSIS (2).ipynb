```python
b=a.drop(['airline','airline_sentiment_confidence','tweet_created','tweet_id'],axis=1)
b
```

```python
b['target'].replace('positive',1,inplace=True)
b['target'].replace('neutral',0,inplace=True)
b['target'].replace('negative',-1,inplace=True)
b['target']=b['target'].astype(int)
b
```

```python
import re
import nltk
# nltk.download('stopwards')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
corpus=[]
for i in range(0, len(b)):
    review=re.sub('[^a-zA-Z]', ' ',b['text'][i])
    review=review.lower()
       # review=review.split()
    review=nltk.word_tokenize(review)
    review= [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review= ' '.join(review)
    corpus.append(review)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500)
x=cv.fit_transform(corpus).toarray()
# y=pd.get_dummies(messages['label'])
y=b.target
```

```python
tweets=b.text
tweets
```

# TOKENIZATION

```python
from nltk import TweetTokenizer
tweetzer=TweetTokenizer()
tweets=tweets.apply(lambda x: tweetzer.tokenize(x)).apply(lambda x: " ".join(x))
tweets
```

# REMOVING SPECIAL CHARACTERS

```python
Llama=tweets.str.replace('[^a-zA=Z0-9]+',' ')
Llama
```

```python
from nltk.tokenize import word_tokenize
Llama2=Llama.apply(lambda x: ' '.join([w for w in word_tokenize(x) if len(w) >=3]))
Llama2
```

# STEMMING

```python
from nltk.stem import SnowballStemmer
stemmer=SnowballStemmer('english')
Llama_rag=Llama2.apply(lambda x:[stemmer.stem(i.lower())for i in tweetzer.tokenize(x)]).apply(lambda x: ' '.join(x))
Llama_rag
```

# STOPWORDS REMOVING

```python
from nltk.corpus import stopwords
sw=stopwords.words('english')
Llama_rag2=Llama_rag.apply(lambda x:[i for i in tweetzer.tokenize(x) if i not in sw]).apply(lambda x: ' '.join(x))
Llama_rag2
```

# VECTORIZATION - TF - IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
turbo=TfidfVectorizer()
fusion_train=turbo.fit_transform(Llama_rag2)
fusion_train.shape
```

```python
fusion_train
```

```python
y=b['target'].values
y
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(fusion_train,y,test_size=0.3,random_state=42)
train_x
```

# K-NEIGHBORS

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(train_x,train_y)
rog3=knn.predict(test_x)
rog3
```

# NAIVE-BAYES

```python
from sklearn.naive_bayes import BernoulliNB
hydra=BernoulliNB()
hydra.fit(train_x,train_y)
rog4=hydra.predict(test_x)
rog4
```

# SVC

```python
from sklearn.svm import SVC
scala=SVC()
scala.fit(train_x,train_y)
rog5=scala.predict(test_x)
rog5
```

# ACCURACY SCORE

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print('Score is:',accuracy_score(test_y,rog3))
```

```python
print('Score is:',accuracy_score(test_y,rog4))
```

```python
print('Score is:',accuracy_score(test_y,rog5))
```
